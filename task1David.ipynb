```python
from selenium import webdriver

driver = webdriver.Chrome()

driver.get("https://es.uefa.com/uefachampionsleague/season=2020/statistics/round=2001140/players")

nextButton = driver.find_element_by_css_selector('.nextLink  .paginate_button')
while nextButton.get_attribute("class").find('paginate_button_disabled') < 0:
    information = driver.find_elements_by_css_selector('td.l span.limg a.js-data-player')
    for item in information:
        print(item.text)    
    nextButton.click()
    nextButton = driver.find_element_by_css_selector('.nextLink  .paginate_button')

## Jugadores de la última página
information = driver.find_elements_by_css_selector('td.l span.limg a.js-data-player')
for item in information:
    print(item.text)    

driver.close()
```
