```python
from selenium.webdriver import Chrome
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
driver = Chrome()
driver.get("https://www.uefa.com/uefachampionsleague/clubs/50051--real-madrid/squad/")
nPlayers = len(driver.find_elements_by_css_selector("a.player-name"))
print("number of players: " + str(nPlayers))
for currentPlayer in range(nPlayers):
    cssSelector = "a.player-name"
    player = driver.find_elements_by_css_selector(cssSelector)
    #print(str(currentPlayer+1) + " Player: " + player[currentPlayer].text)
    time.sleep(2)
    #player[currentPlayer].click()
    driver.execute_script("arguments[0].click();", player[currentPlayer])
    time.sleep(2)
    
    total = {}
    statisticsListBlock = driver.find_elements_by_css_selector("div.player--statistics--list")
    if(len(statisticsListBlock) == 1):
        fields = driver.find_elements_by_css_selector("div.player--statistics--list .field")
        used=[] 
        class Player(object):
            def __init__(self, id, name, team, goals, saves, offsides, minutesPlayed, goalsConceded, played):
                self.id=id
                self.name=name
                self.team=team
                self.goals=goals
                self.saves=saves
                self.offsides=offsides
                self.minutesPlayed=minutesPlayed
                self.goalsConceded=goalsConceded
                self.played=played
        for field in fields:
            label = field.find_element_by_class_name("statistics--list--label").text
            data = field.find_element_by_class_name("statistics--list--data").text
            total.update({label: data})
        url = driver.current_url
        x=url.split('/')
        data=x[6].split('--')
        player_id=data[0]
        name=data[1]
        team=driver.find_element_by_class_name("player-header_team-name").text
        player = Player(player_id, name, team, total["GOALS"], total["SAVES"], total["OFFSIDES"], total["MINUTES PLAYED"], total["GOALS CONCEDED"], total["PLAYED"])
        print("id: " + player.id + " , -Name: " + player.name + " , Team: " + player.team + " , Goals: " + player.goals + " , Saves: " + player.saves + " , Offsides: " + player.offsides + " , Minutes Played: " + player.minutesPlayed + " , Goals Conceded: " + player.goalsConceded + " ,Played: " + player.played)
    
    else:
        name = driver.find_element_by_class_name("player-header_name").text
        print(name + " has not played any matches")
    
    driver.execute_script("window.history.go(-1)")
    
driver.close()
```
